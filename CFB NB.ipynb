```python
def get_rank(teamStr):
    match = re.search('\([0-9]*\) ',teamStr)
    if match: 
        return match.group().lstrip('(').rstrip(') ')
    

    
```

```python
#read in games data and teams by conference
cfb = pd.read_csv('cfb-data.csv')
conf = pd.read_csv('CFB-teams-by-conference.csv')

def get_conf(team, df = conf):
    s =df[df.Team == team]
    s.reset_index()
    if not s.empty:
        return s.Conf.values[0]
    else:
        return 'FCS'

#tidy up data...remove repeated headers and convert strings to floats for Pts columns
cfb = cfb[(cfb['Rk'] != 'Rk')]
cfb[['Pts', 'Pts.1']] = cfb[['Pts', 'Pts.1']].astype(float)

#remove NAs... i.e. games not yet played
cfb = cfb[np.isfinite(cfb['Pts'])]

#tidy up columns
cfb.drop('Unnamed: 12', axis=1, inplace=True)
cfb.rename(columns={'Unnamed: 7':'home_or_away', 'Winner/Tie':'Winner', 'Loser/Tie':'Loser'}, inplace=True)
cfb['home_lost'] = (cfb['home_or_away'] == "@").astype(int)

#get rankings at time of game
cfb['winner_rank'] = cfb['Winner'].apply(get_rank)
cfb['loser_rank'] = cfb['Loser'].apply(get_rank)


#remove ranks from team names
cfb['Winner'] = cfb['Winner'].map(lambda x: re.sub(r'\([0-9]*\) ','',x))
cfb['Loser'] = cfb['Loser'].map(lambda x: re.sub(r'\([0-9]*\) ','',x))

#add conference for each team
cfb['winner_conf'] = cfb['Winner'].apply(get_conf)
cfb['loser_conf'] = cfb['Loser'].apply(get_conf)

```

```python
#sum non conf wins for each conference
def set_non_conf_record(curConf, df1 = cfb, df2 = conf, ignore_fcs = False, power_5_only = False):
    #using a global probably not the best way
    global conf_records
    power_5 =['SEC', 'Pac-12','Big Ten', 'ACC', 'Big 12']
    winRows = df1[(df1.winner_conf == curConf) & (df1.loser_conf != curConf)]
    lossRows = df1[(df1.loser_conf == curConf) & (df1.winner_conf != curConf)]
    
    #ignore wins vs FCS schools; losses still count
    if ignore_fcs:
        winRows = winRows[(winRows['loser_conf'] != 'FCS')]
        
    if power_5_only:
        winRows = winRows[winRows['loser_conf'].isin(power_5)]
        lossRows = lossRows[lossRows['winner_conf'].isin(power_5)]

    
    numWins = len(winRows)
    numLosses = len(lossRows)
    wpct = numWins/(numWins + numLosses)
    win_pts_diff = winRows.Pts.sum()-winRows['Pts.1'].sum()
    loss_pts_diff = lossRows['Pts.1'].sum()-lossRows['Pts'].sum()
    pts_diff = win_pts_diff + loss_pts_diff
    data = {'Conf': curConf, 'wins': numWins,'losses': numLosses,'wpct':wpct, 'pts_diff':pts_diff}
    conf_row = pd.Series(data=data)
    conf_records = conf_records.append(conf_row, ignore_index=True)
    conf_records.drop_duplicates(subset=('Conf'), inplace=True)


conf_records = pd.DataFrame()
conf['Conf'].apply(set_non_conf_record, df2 = conf_records)
non_conf_total = conf_records

conf_records = pd.DataFrame()
conf['Conf'].apply(set_non_conf_record, df2 = conf_records, ignore_fcs = True)
non_conf_no_fcs = conf_records

conf_records = pd.DataFrame()
conf['Conf'].apply(set_non_conf_record, df2 = conf_records, power_5_only = True)
non_conf_power_5 = conf_records

def plotWvL(df, width = 1, lossColor = 'r', winColor = 'g', title = 'Wins by Conference'):
    
    p1 = plt.barh(np.arange(len(df.wins)),df.wins + df.losses , width, color=lossColor)
    p2 = plt.barh(np.arange(len(df.losses)),df.wins, width, color=winColor)
    plt.ylabel('Conference')
    
    plt.title(title)
    plt.yticks(np.arange(len(df.wins)) + width/2, df.Conf )
    plt.legend( (p1[0], p2[0]), ('Losses', 'Wins'),bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
    

```

### Total Non-Conference Wins and Losses

```python
print(non_conf_total.sort(columns='wpct', ascending=False))
plotWvL(non_conf_total, 0.75, winColor = 'b', lossColor = 'r', title = 'Total Non Conference Wins')
```

### Non-Conference Wins and Losses Without FCS Opponents

```python
print(non_conf_no_fcs.sort(columns='wpct', ascending=False))
plotWvL(non_conf_no_fcs, 0.75, winColor = 'y', lossColor = 'r', title = 'Non Conference Wins Without FCS Opponents')
```

### Non-Conference Wins and Losses Vs Power 5 (SEC, Big Ten, Pac 12, Big 12, ACC)

```python
print(non_conf_power_5.sort(columns='wpct', ascending=False))
plotWvL(non_conf_power_5, 0.75, winColor = 'g', lossColor = 'r', title = 'Non Conference Power 5 Wins by Conference')
```

```python

width = 1.0    
p1 = plt.barh(np.arange(len(non_conf_total.wins)),non_conf_total.wins , width, color='b')
p2 = plt.barh(np.arange(len(non_conf_total.wins)),non_conf_no_fcs.wins, width, color='y')
p3 = plt.barh(np.arange(len(non_conf_total.wins)),non_conf_power_5.wins, width, color='g')
plt.ylabel('Conference')
plt.xlabel('Wins')
plt.title('Wins by Conference by Opponent')
plt.yticks(np.arange(len(non_conf_total.wins)) + width/2, non_conf_power_5.Conf )
plt.legend( (p1[0], p2[0], p3[0]), ('FCS Wins', 'Non-Power 5 Wins', 'Power 5 Wins'), bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0. )
plt.show()
```
